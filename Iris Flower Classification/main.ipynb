Graphs

```python
colors = ['deepskyblue', 'mediumseagreen', 'orchid']  # More vibrant colors
species = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']

sns.set_style('ticks')  # Set a visually appealing style

for i in range(3):
    x = df[df['Species'] == species[i]]
    plt.scatter(x['SepalLengthCm'], x['SepalWidthCm'],
                c=colors[i], label=species[i], s=75, edgecolors='black', linewidth=0.5)  # Adjust marker size and add subtle edge colors

plt.xlabel('Sepal Length (cm)', fontsize=12)  # Increase font size for readability
plt.ylabel('Sepal Width (cm)', fontsize=12)
plt.title('Sepal Length vs. Width by Iris Species', fontsize=14)  # Add a descriptive title
plt.legend(fontsize=10)  # Adjust legend font size
plt.grid(True, alpha=0.5)  # Add a light grid for better data interpretation
plt.xlim(4, 8)  # Adjust x-axis limits for better visualization
plt.ylim(2, 5)  # Adjust y-axis limits for better visualization
plt.show()
```

Coorelation Matrix

```python
corr = df.select_dtypes(include=[np.number]).corr() #not including the species column
print(corr)
```

```python
corr = df.select_dtypes(include=[np.number]).corr()  # Calculate correlation
sns.set_style('whitegrid')  # Set a visually appealing style
fig, ax = plt.subplots(figsize=(7, 5))  # Adjust figure size
cmap = sns.diverging_palette(220, 10, as_cmap=True)  # Select a diverging colormap

sns.heatmap(corr, annot=True, cmap=cmap, fmt='.2f', square=True, linewidths=0.5, cbar_kws={'shrink': 0.8}, ax=ax)
plt.title('Correlation Matrix for Numerical Features', fontsize=14)  # Add a descriptive title
plt.show()
```

Label Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
df['Species'] = le.fit_transform(df['Species'])
df.head()
```

Model training

```python
from sklearn.model_selection import train_test_split

X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.7)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print('Accuracy using Logistic Regression: ', model.score(x_test, y_test)*100)
```

Using KNN (K-Nearest Neighbour)

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier()

model.fit(x_train, y_train)
```

```python
print('Accuracy using KNN: ', model.score(x_test, y_test)*100)
```

Using Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()

model.fit(x_train, y_train)
```

```python
print('Accuracy using Decision Tree: ', model.score(x_test, y_test)*100)
```
